## **Preprocessing**

````python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from sklearn.preprocessing import MinMaxScaler

class DataPreprocessing:

    # constructor method to initialize the DataPreprocessing class with participant timings data
    def __init__(self, participant_timings):
        self.participant_timings = participant_timings

    # convert a time string (participant timings) in the format 'minutes.seconds' to seconds
    def convert_to_seconds(self, time_str):
        minutes, seconds = map(int, time_str.split('.'))
        return (minutes * 60) + seconds

    # assigns a label based on the provided timestamp and state timings.
    def assign_rating(self, timestamp, state_timings):
        assigned_rating = 0  # default label is 0

        # convert the timestamp to hours, minutes, and seconds
        hours = timestamp // 3600
        timestamp %= 3600
        minutes = timestamp // 60
        seconds = timestamp % 60
        timestamp_seconds = hours * 3600 + minutes * 60 + seconds

        current_state = None
        # iterate through the state timings and assign a label if the timestamp falls within a range
        for state, timings in state_timings.items():
            start_time = self.convert_to_seconds(timings["start"])
            end_time = self.convert_to_seconds(timings["end"])
            if start_time <= timestamp_seconds <= end_time:
                current_state = state
                # exit the loop once a label is assigned
                break

        # If a state with a rating is found, use that rating; otherwise, use the default rating (0)
        if current_state:
            assigned_rating = state_timings[current_state]["rating"]

        return assigned_rating

    # preprocessing for individual participants
    def preprocess_participant(self, participant):

        # access participant file path to CSV
        hr_file_path = f'/content/drive/MyDrive/WESAD/S{participant}/S{participant}_E4_Data/HR.csv'
        eda_file_path = f'/content/drive/MyDrive/WESAD/S{participant}/S{participant}_E4_Data/EDA.csv'

        # load data from CSV files into dataframes, excluding the first two rows which contain timestamp and Hz information
        hr_df = pd.read_csv(hr_file_path, header=None, skiprows=2)
        eda_df = pd.read_csv(eda_file_path, header=None, skiprows=2)

        # get the state timings for the current participant
        state_timings = self.participant_timings.get(participant, {})


        ## SAMPLING RATE PROCESS

        # add a timestamp column based on the provided sampling rate
        sampling_rate_hr = 1
        sampling_rate_eda = 4
        hr_df['Timestamp'] = pd.date_range(start=0.0, periods=len(hr_df), freq=str(1 / sampling_rate_hr) + 'S')
        eda_df['Timestamp'] = pd.date_range(start=0.0, periods=len(eda_df), freq=str(1 / sampling_rate_eda) + 'S')

        # rename readings column to appropriate names
        hr_df = hr_df.rename(columns={0: 'HeartRateData'})
        eda_df = eda_df.rename(columns={0: 'EDAData'})

        hr_df.set_index('Timestamp', inplace=True)
        eda_df.set_index('Timestamp', inplace=True)


        ## TIME ALIGNMENT PROCESS

        # determine the common start time
        common_start_time = max(hr_df.index.min(), eda_df.index.min())

        # determine the common end time
        common_end_time = min(hr_df.index.max(), eda_df.index.max())

        # trim HR and EDA data to match the common start and end times
        hr_df = hr_df.truncate(before=common_start_time, after=common_end_time)
        eda_df = eda_df.truncate(before=common_start_time, after=common_end_time)


        ## RESAMPLING PROCESS  (UPSAMPLING)

        # resample HR data to 4 Hz
        hr_df = hr_df.resample('0.25S').ffill()

        # add a new column named 'Rating' to hr_df_resampled based on the assign_rating function (state timings)
        hr_df['Rating'] = hr_df.index.map(lambda x: self.assign_rating(x.timestamp(), state_timings))
        eda_df['Rating'] = eda_df.index.map(lambda x: self.assign_rating(x.timestamp(), state_timings))

        # Filter out rows with a rating of 0 after reindexing
        hr_df = hr_df[hr_df['Rating'] != 0]
        eda_df = eda_df[eda_df['Rating'] != 0]

        # reset index from Timestamp to integer
        hr_df.reset_index(drop=True, inplace=True)
        eda_df.reset_index(drop=True, inplace=True)

        return hr_df, eda_df
````

````python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

preprocessor = DataPreprocessing(participant_timings)
hr_df_list = []
eda_df_list = []

for participant in participant_timings.keys():
    hr_preprocessed_data, eda_preprocessed_data = preprocessor.preprocess_participant(participant)
    hr_df_list.append(hr_preprocessed_data)
    eda_df_list.append(eda_preprocessed_data)

hr_preprocessed_data = pd.concat(hr_df_list, ignore_index=True)
eda_preprocessed_data = pd.concat(eda_df_list, ignore_index=True)

hr_preprocessed_data = hr_preprocessed_data.sample(frac=1, random_state=42)
eda_preprocessed_data = eda_preprocessed_data.sample(frac=1, random_state=42)
````

````python
import numpy as np
import skfuzzy as fuzz
import random as rd
from skfuzzy import control as ctrl
from skfuzzy import membership as mf
import sys
````

````python
# Class for fuzzy system
class FuzzySystem:

  def __init__(self):

    # Initialize inputs and outputs
    self.heart_rate = ctrl.Antecedent(np.arange(50, 160, 0.1), 'HeartRateData')
    self.EDA = ctrl.Antecedent(np.arange(0, 16, 0.001), 'EDAData')
    self.rating = ctrl.Consequent(np.arange(1, 5, 0.1), 'Rating')

    # Define membership functions for fuzzy sets of variables
    self.heart_rate['low'] = mf.trapmf(self.heart_rate.universe, [50, 50, 62.1, 74.02])
    self.heart_rate['medium'] = mf.trapmf(self.heart_rate.universe, [64.03, 70.03, 79.97, 88.98])
    self.heart_rate['high'] = mf.trapmf(self.heart_rate.universe, [79.0, 113.79, 160, 160])

    self.EDA['low'] = mf.trapmf(self.EDA.universe, [0, 0, 0.319, 0.378])
    self.EDA['medium'] = mf.trimf(self.EDA.universe, [0.278, 0.4, 2.936])
    self.EDA['high'] = mf.trapmf(self.EDA.universe, [2.387, 5, 16, 16])

    self.rating['low'] = mf.trimf(self.rating.universe, [1, 1, 3])
    self.rating['medium'] = mf.trimf(self.rating.universe, [2, 3, 4])
    self.rating['high'] = mf.trimf(self.rating.universe, [3, 5, 5])

    # Define rules
    rule1 = ctrl.Rule(self.heart_rate['low'] & self.EDA['low'], self.rating['medium'])
    rule2 = ctrl.Rule(self.heart_rate['low'] & self.EDA['medium'], self.rating['medium'])
    rule3 = ctrl.Rule(self.heart_rate['low'] & self.EDA['high'], self.rating['medium'])
    rule4 = ctrl.Rule(self.heart_rate['medium'] & self.EDA['low'], self.rating['medium'])
    rule5 = ctrl.Rule(self.heart_rate['medium'] & self.EDA['medium'], self.rating['low'])
    rule6 = ctrl.Rule(self.heart_rate['medium'] & self.EDA['high'], self.rating['high'])
    rule7 = ctrl.Rule(self.heart_rate['high'] & self.EDA['low'], self.rating['medium'])
    rule8 = ctrl.Rule(self.heart_rate['high'] & self.EDA['medium'], self.rating['low'])
    rule9 = ctrl.Rule(self.heart_rate['high'] & self.EDA['high'], self.rating['high'])

    self.rules = [rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8, rule9]

    # Create a control system based on the defined rules
    self.rating_ctrl = ctrl.ControlSystem(self.rules)

  def showMembershipGraph(self):
    self.heart_rate.view()
    self.EDA.view()
    self.rating.view()

  def fuzzyInferencing(self, input_heart_rate, input_eda):

    # Create a control system simulation
    rating_sim = ctrl.ControlSystemSimulation(self.rating_ctrl)

    # Input specific values
    rating_sim.input['HeartRateData'] = input_heart_rate
    rating_sim.input['EDAData'] = input_eda

    # Perform fuzzy inference
    rating_sim.compute()

    return rating_sim

  def getMembershipValues(self, input_heart_rate, input_eda, rating_sim):

    # Fuzzify the inputs
    heart_rate_membership_low = fuzz.interp_membership(self.heart_rate.universe, self.heart_rate['low'].mf, input_heart_rate)
    heart_rate_membership_medium = fuzz.interp_membership(self.heart_rate.universe, self.heart_rate['medium'].mf, input_heart_rate)
    heart_rate_membership_high = fuzz.interp_membership(self.heart_rate.universe, self.heart_rate['high'].mf, input_heart_rate)

    eda_membership_low = fuzz.interp_membership(self.EDA.universe, self.EDA['low'].mf, input_eda)
    eda_membership_medium = fuzz.interp_membership(self.EDA.universe, self.EDA['medium'].mf, input_eda)
    eda_membership_high = fuzz.interp_membership(self.EDA.universe, self.EDA['high'].mf, input_eda)

    # Access the degree of membership for each linguistic label
    degree_low_stress = fuzz.interp_membership(self.rating.universe, self.rating['low'].mf, rating_sim.output['Rating'])
    degree_medium_stress = fuzz.interp_membership(self.rating.universe, self.rating['medium'].mf, rating_sim.output['Rating'])
    degree_high_stress = fuzz.interp_membership(self.rating.universe, self.rating['high'].mf, rating_sim.output['Rating'])

    # Display the membership values
    print("Heart Rate Memberships (Low, Medium, High):", heart_rate_membership_low, heart_rate_membership_medium, heart_rate_membership_high)
    print("EDA Memberships (Low, Medium, High):", eda_membership_low, eda_membership_medium, eda_membership_high)
    print("Stress Level Memberships (Low, Medium, High):", degree_low_stress, degree_medium_stress, degree_high_stress)

  def getAreaUnderCurve(self, rating_sim):

    # Visualize the membership functions and area under the curve
    self.rating.view(sim=rating_sim)

    plt.show()

  def defuzzify_output(self, rating_sim):

    # Defuzzify the output
    defuzzified_value = rating_sim.output['Rating']

    return defuzzified_value

  def recommendation(self, heart_rate, eda):

    # Provide recommendations
    activities = [["You are doing good!"],
                  ["Practice mindfulness meditation to improve mental clarity",
                   "Write down your thoughts somewhere to express your feelings.",
                   "Talk to a friend"],
                  ["Take a short break and try to catch some breaths",
                   "Consider consulting a mental health professional for counseling or therapy",
                   "Engage in stress-reducing activities you enjoy.",
                   "Reduce exposure to stress-inducing news or social media by limiting screen time."]]

    # Simulate ratings
    rating_sim = ctrl.ControlSystemSimulation(self.rating_ctrl)
    rating_sim.input['HeartRateData'] = heart_rate
    rating_sim.input['EDAData'] = eda
    rating_sim.compute()

    # Determine the stress level
    x = rating_sim.output['Rating']

    if x < 2:
        return activities[0][0]  # "low" stress recommendation
    elif x < 3:
        return activities[1][rd.randint(0, 2)]  # "medium" stress recommendation
    else:
        return activities[2][rd.randint(0, 3)]  # "high" stress recommendation
````
